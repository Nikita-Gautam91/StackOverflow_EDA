```python
numerical_columns = survey_raw_df.select_dtypes(include=['int','float']).columns
survey_raw_df[numerical_columns] = survey_raw_df[numerical_columns].fillna(survey_raw_df[numerical_columns].mean())
```

```python
categorical_columns = survey_raw_df.select_dtypes(include=['object']).columns
survey_raw_df[categorical_columns] = survey_raw_df[categorical_columns].fillna(survey_raw_df[categorical_columns].mode().iloc[0])
```

```python
survey_raw_df.isnull().sum()
```

```python
selected_columns = [
    #Basic Information
    'ResponseId',
    'MainBranch',
    'Age',
    'Employment',
    'RemoteWork',
    #Education, Work, and Career
    'EdLevel',
    'YearsCode',
    'YearsCodePro',
    'DevType',
    'Country',
    'CompTotal',
    #Technology and Tech Culture
    'LanguageHaveWorkedWith',
    'LanguageWantToWorkWith',
    'DatabaseHaveWorkedWith',
    'DatabaseWantToWorkWith',
    'PlatformHaveWorkedWith',
    'PlatformWantToWorkWith',
    'WebframeHaveWorkedWith',
    'WebframeWantToWorkWith',
    'OpSysPersonal use',
    'OpSysProfessional use',
    #Stack oveerflow
    'SOVisitFreq',
    # AI
    'AIToolInterested in Using',
    'AIToolCurrently Using',
    'AISent',
    'AIAcc',
    'AIBen'
] 
```

```python
survey_df = survey_raw_df[selected_columns].copy()
```

```python
survey_df.shape
```

```python
survey_df.info()
```

```python
survey_df.describe()
```

```python
survey_df['Age'] = pd.to_numeric(survey_df['Age'], errors='coerce')
survey_df['YearsCode'] = pd.to_numeric(survey_df['YearsCode'], errors='coerce')
survey_df['YearsCodePro'] = pd.to_numeric(survey_df['YearsCodePro'], errors='coerce')
```

```python
survey_df.describe()
```

```python
survey_df['Age']
```

```python
age_mapping = {
    'Under 18 years old' : 16,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70,
    'Prefer not to say': -1,
}
```

```python
survey_raw_df['NumericalAge']= survey_raw_df['Age'].map(age_mapping)
```

# Top ten countries with the highest number of responses in the survey

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
country_counts = survey_raw_df['Country'].value_counts()
Sorting = country_counts.sort_values(ascending=False)
Top_10_countries = Sorting.head(10)
Top_10_countries 
```

```python
plt.figure(figsize=(10, 6))
Top_10_countries.plot(kind='bar')
plt.title('Top 10 Countries by Survey Responses')
plt.xlabel('Country')
plt.ylabel('Number of Responses')
plt.xticks(rotation=45)
plt.show()
```

# How much do you trust the accuracy of the output from AI tools as part of your development workflow?

```python
cleaned_ai_ben_data = survey_raw_df['AIBen'].dropna()
accuracy_counts = cleaned_ai_ben_data.value_counts()
custom_palette = ["#ff9999", "#66b3ff", "#99ff99","#55a868", "#c44e52"]
plt.figure(figsize=(10, 6))
sns.barplot(x=accuracy_counts.index, y=accuracy_counts.values, palette=custom_palette)
plt.title('Perceived Accuracy of AI Tools', fontsize=16)
plt.xlabel('Accuracy Level', fontsize=14)
plt.ylabel('Number of Respondents', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

# For the AI tools you use as part of your development workflow, what are the MOST important benefits you are hoping to achieve

```python
cleaned_ai_acc_data = survey_raw_df['AIAcc'].dropna()
mapping_dict = {
    'Increase productivity;Greater efficiency;Speed up learning;Improve accuracy in coding': 'All',
    'Greater efficiency': 'Efficiency',
    'Improve accuracy in coding': 'Accuracy',
    'Increase productivity': 'Productivity',
    'Speed up learning': 'Learning',
    'Other (please explain)': 'Other'
}
cleaned_ai_acc_data = cleaned_ai_acc_data.map(mapping_dict)
mapped_benefit_counts = cleaned_ai_acc_data.value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=mapped_benefit_counts.index, y=mapped_benefit_counts.values, palette='pastel')
plt.title('Perceived Benefits of AI Tools', fontsize=16)
plt.xlabel('Benefit Category', fontsize=14)
plt.ylabel('Number of Respondents', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```
